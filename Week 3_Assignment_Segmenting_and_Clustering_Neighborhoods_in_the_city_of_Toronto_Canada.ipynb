### Load packages

```python
from bs4 import BeautifulSoup
import requests
import csv
```

### Scape information from the wiki page and save it into a csv file

```python
csv_file = open('toronto_scrape.csv', 'w', newline="")
csv_writer = csv.writer(csv_file)

source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')
table = soup.find('table', class_= 'wikitable sortable').tbody

for each_tr in table.find_all('tr'):
    temp = []
    for each_th in each_tr:
        try:
            temp.append(each_th.text.strip())
        except Exception as e:
            continue
           
    print(temp)
    csv_writer.writerow(temp)
csv_file.close()
```

### Read the save csv file 

```python
import pandas as pd
data = pd.read_csv("toronto_scrape.csv")
data
```

### Get rid of rows that have "Borough" values equal to "Not assigned"

```python
data_new = data[data["Borough"]!="Not assigned"]
data_new
```

### Combine rows with same values of "Postcode" and "Borough"

```python
df = pd.DataFrame(data_new.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index())
df
```

### Check whether there is any row which has a borough but a "Not assigned" neighborhood

```python
sum(df['Neighbourhood']=='Not assigned')

# Note: It turned out there is no such rows, so we don't need to assign the values of Borough to Neighbourhood. 
```

### Find the latitude and longitude of Toronto

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="can_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
